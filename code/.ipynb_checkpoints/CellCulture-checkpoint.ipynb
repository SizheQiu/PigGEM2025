```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
from scipy import stats
```

```python
def add_linear(ax, xdata, ydata):
    def func(x, a, b):
        return a * x + b
    popt, pcov = curve_fit(func, xdata, ydata)
    temp_x = np.linspace(min(xdata),max(xdata) )
    ax.plot(temp_x, func(temp_x, *popt),'r--')
    
def get_pv(x,y):
    coeff, pv = stats.pearsonr( x, y)
    return pv

def get_pcc(x,y):
    coeff, pv = stats.pearsonr( x, y)
    return coeff
```

```python
data=pd.read_csv('../data/exp_data/CultureData.csv')
data=data.rename(columns={'Biomass_avg(g/L)':'Biomass_avg','Biomass_std(g/L)':'Biomass_std'})
```

```python
fig, axs = plt.subplots(2,3,figsize=(16,8))
ynames=['Biomass','Glucose','Gln','Lac','NH4']
for i in range(5):
    x,y=int(i/3),int(i%3)
    axs[x][y].errorbar(data.Time,data[ynames[i]+'_avg'],yerr= data[ynames[i]+'_std'], ecolor='black',
                      marker='o', markersize=5, linestyle='-', linewidth=0.5, elinewidth=1)
    axs[x][y].set_xlabel('Time(hr)')
    axs[x][y].set_ylabel(ynames[i]+'(mM)')
fig.delaxes(axs[1][2])
axs[0][0].set_ylabel('Biomass(gDW/L)');
```

```python
def get_grs(biomass, time_pts):
    output = []
    for i in range(len(biomass)-1):
        output.append( np.log(biomass[i+1]/biomass[i]) /(time_pts[i+1]-time_pts[i]) )
    return output  

def norm_rates(biomass, cons, time_pts):
    output = []
    for i in range(len(biomass)-1):
        output.append( ((cons[i+1]-cons[i])/(biomass[i]/2+biomass[i+1]/2) )/(time_pts[i+1]-time_pts[i]) )
    return output
```

```python
Ts= [data.Time[i]/2+data.Time[i+1]/2 for i in range(len(data.Time)-1)]
grs=get_grs(data.Biomass_avg, data.Time)
growth_data = pd.DataFrame(zip(Ts,grs),columns=['Time','gr'])
mets=['Glucose','Gln','Lac','NH4']
for met in mets:
    growth_data[met+'_rate']=norm_rates( data.Biomass_avg, data[met+'_avg'], data.Time)
```

```python
fig,axs = plt.subplots(1,3,figsize=(15,4))
axs[0].plot(Ts,growth_data.gr,marker='o', markersize=5, linestyle='--')
axs[0].set_ylabel('Growth rate(1/hr)')
axs[1].plot(Ts,-1*growth_data.Lac_rate/growth_data.Glucose_rate,marker='o', markersize=5, linestyle='--')
axs[1].set_ylabel('Lactate/Glucose rate')
axs[2].plot(Ts,-1*growth_data.NH4_rate/growth_data.Gln_rate,marker='o', markersize=5, linestyle='--')
axs[2].set_ylabel('NH4+/Glutamine rate')
```

```python
ATP_approx=[]
for i in range(len(growth_data.index)):
    lac_r=growth_data.Lac_rate[i];glc_r=(-1)*growth_data.Glucose_rate[i];gln_r=(-1)*growth_data.Gln_rate[i];
    ATP_approx.append( lac_r + (glc_r-lac_r/2)*38+gln_r*24 )
growth_data['ATP_estimated']=ATP_approx
```

```python
def func(x, a, b):
    return a * x + b
popt, pcov = curve_fit(func, growth_data.gr[1:],growth_data.ATP_estimated[1:])
print(popt)
```

```python
fig,ax=plt.subplots(1,1,figsize=(5,5))
ax.scatter(growth_data.gr,growth_data.ATP_estimated,color='blue')
ax.scatter(growth_data.gr[1:],growth_data.ATP_estimated[1:],color='orange')
ax.set_ylabel('ATP generated')
ax.set_xlabel('Growth rate(1/hr)')
temp_x = np.linspace(0,0.07)
ax.plot(temp_x, func(temp_x, *popt),'r--')
y_pred = [func(x,176.53898486, 5.06580435) for x in growth_data.gr[1:]]
print( get_pv(growth_data.ATP_estimated[1:],y_pred) )
print( get_pcc(growth_data.ATP_estimated[1:],y_pred) )
```
